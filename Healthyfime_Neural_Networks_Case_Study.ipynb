```python
X=df.iloc[:,:-1]
```

```python
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split


X_dev,X_test,y_dev,y_test=train_test_split(X,Y,test_size=0.1,random_state=42)
X_train,X_val,y_train,y_val=train_test_split(X_dev,y_dev,test_size=0.1,random_state=42)

```

```python
print(X_train.shape,y_train.shape)
```

```python
print(X_val.shape,y_val.shape)
```

```python
print(X_test.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_val=scaler.transform(X_val)
X_test=scaler.transform(X_test)
```

```python
y_train
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

#here we are defining  straight away the first hidden layer because we are using a sequential class object so by default the input layer has to be fed in the firt hidden layer

model.add(Dense(20,input_dim=X_train.shape[1],activation='tanh'))


model.add(Dense(10,activation='relu'))
model.add(Dense(4,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=50)
```

```python
model.fit(X_test,y_test,epochs=100)
```
